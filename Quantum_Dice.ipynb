```python
from qiskit import *
from qiskit import transpile
from qiskit.providers.aer import AerSimulator
```

```python
def menu():
    print("1. QD2")
    print("2. QD4")
    print("3. QD6")
    print("4. QD8")
    print("5. QD10")
    print("6. QD12")
    print("7. QD20")
    print("8. QD100")
    print("0. Exit")

    while True:
        print("-"*80)
        choice = input("Selection: ")

        if choice == '1':
            QD2()
        elif choice == '2':
            QD4()
        elif choice == '3':
            QD6()
        elif choice == '4':
            QD8()
        elif choice == '5':
            QD10()
        elif choice == '6':
            QD12()
        elif choice == '7':
            QD20()
        elif choice == '8':
            QD100()
        elif choice == '0':
            exit(0)
        else:
            print("Invalid selection")

def QD2():
    print("\nQD2\n")
    qr = QuantumRegister(1, 'q')
    cr = ClassicalRegister(1, 'c')
    qc = QuantumCircuit(qr,cr)

    qc.h(qr)
    qc.measure(qr,cr)

    backend = AerSimulator()
    qc_compiled = transpile(qc, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)

    cmf = counts.most_frequent()

    if cmf == '0':
        print("Heads")
    elif cmf == '1':
        print("Tails")

def QD4():
    qr = QuantumRegister(2)
    cr = ClassicalRegister(2)
    qc = QuantumCircuit(qr,cr)

    qc.h(qr)
    qc.measure(qr,cr)
    
    backend = AerSimulator()
    qc_compiled = transpile(qc, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)

    cmf = counts.most_frequent()
    print("\nQD4\n")
    print("You rolled: ", int(cmf,2)+1)

def QD6():
    qr = QuantumRegister(3)
    cr = ClassicalRegister(3)
    qc = QuantumCircuit(qr, cr)

    qc.h(qr)
    qc.measure(qr,cr)
    
    backend = AerSimulator()
    qc_compiled = transpile(qc, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)

    cmf = counts.most_frequent()

    if cmf <= '101':
        print("\nQD6\n")
        print("You rolled: ", int(cmf,2)+1)
    elif cmf > '101':
        QD6()


def QD8():
    qr = QuantumRegister(3)
    cr = ClassicalRegister(3)
    qc = QuantumCircuit(qr,cr)

    qc.h(qr)
    qc.measure(qr,cr)

    backend = AerSimulator()
    qc_compiled = transpile(qc, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)

    cmf = counts.most_frequent()
    print("\nQD8\n")
    print("You rolled: ", int(cmf,2)+1)

def QD10():
    qr = QuantumRegister(4)
    cr = ClassicalRegister(4)
    qc = QuantumCircuit(qr,cr)

    qc.h(qr)
    qc.measure(qr,cr)

    backend = AerSimulator()
    qc_compiled = transpile(qc, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)

    cmf = counts.most_frequent()

    if cmf <= '1001':
        print("\nQD10\n")
        print("You rolled: ", int(cmf,2)+1)
    elif cmf > '1001':
        QD10()

def QD12():
    qr = QuantumRegister(4)
    cr = ClassicalRegister(4)
    qc = QuantumCircuit(qr,cr)

    qc.h(qr)
    qc.measure(qr,cr)

    backend = AerSimulator()
    qc_compiled = transpile(qc, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)

    cmf = counts.most_frequent()

    if cmf <= '1011':
        print("\nQD12\n")
        print("You rolled: ", int(cmf,2)+1)
    elif cmf > '1011':
        QD12()

def QD20():
    qr = QuantumRegister(5)
    cr = ClassicalRegister(5)
    qc = QuantumCircuit(qr,cr)

    qc.h(qr)
    qc.measure(qr,cr)

    backend = AerSimulator()
    qc_compiled = transpile(qc, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)

    cmf = counts.most_frequent()
    
    if cmf <= '10011':
        print("\nQD20\n")
        print("You rolled: ", int(cmf,2)+1)
    elif cmf > '10011':
        QD20()

def QD100():
    qr = QuantumRegister(7)
    cr = ClassicalRegister(7)
    qc = QuantumCircuit(qr,cr)

    qc.h(qr)
    qc.measure(qr,cr)

    backend = AerSimulator()
    qc_compiled = transpile(qc, backend)
    job_sim = backend.run(qc_compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(qc_compiled)

    cmf = counts.most_frequent()

    if cmf <= '1100011':
        print("\nQD100\n")
        print("You rolled: ", int(cmf,2)+1)
    elif cmf > '1100011':
        QD100()

menu()
```
